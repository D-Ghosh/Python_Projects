```python
import matplotlib.pyplot as plt

def fz(a):
    return a/(2 + 0.0156*(a**(2/3)))
N, Z, i = [], [], 0
for A in range(100):
    Z.append(round(fz(A)))
    N.append(A - Z[i])
    i += 1
N0 = Z
print(N,"\n\n", Z)
plt.scatter(Z, N)
plt.plot(Z, N0)
plt.show()
```
