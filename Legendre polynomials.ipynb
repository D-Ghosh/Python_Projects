```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-1.5, 1.5, 1000)
P0 = x**0
P1 = x
P2 = (1/2)*(3*x**2 - 1)
P3 = (1/2)*(5*x**3 - 3*x)
P4 = (1/8)*(35*x**5 - 30*x**3 + 3)

plt.ylim(-1.2, 1.2)
plt.xlim(-1.2, 1.2)

plt.plot(x, P0, x, P1, x, P2, x, P3, x, P4)

plt.legend(["$P_0$", "$P_1$", "$P_2$", "$P_3$", "$P_4$"])

plt.axhline(ls = "--", c = "k")

plt.title("Legendre Polynomials", size = "20")

plt.show()
```
