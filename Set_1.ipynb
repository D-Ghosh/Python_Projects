```python
# Q1(i)
import numpy as np
import math

x = [np.random.uniform(0, 2) for i in range(50)]
x2 = [i**2 for i in x]

avg_x = sum(x)/len(x)
avg_x2 = sum(x2)/len(x2)

sigma = math.sqrt(avg_x2 - avg_x**2)
print(sigma)
```

```python
# Q1(ii)
# Integration by composite Simpson's rule

import numpy as np
import math

a, b = eval(input("a, b\n"))
k = math.sqrt(1-(a**2/b**2))

def f(x): return 4*b*math.sqrt(1-(k**2)*math.sin(x)**2)

x1, x2, n = eval(input("Lower limit, Upper limit, n\n"))
x2 = x2/2
h = float(x2 - x1)/n
s = f(x1) + f(x2)
d = 4

for i in range(1, n):
    x = x1 + i*h
    s += d*f(x)
    d = 6-d

s = h/3*s
print(s)
```

```python
# Q2(ii)

def f(x): return (x**4)-1.99*(x**3)-1.76*(x**2)-5.22*x-2.33

a, b, tol = eval(input("Lower limit, Upper limit, Tolerance\n"))

if f(a)*f(b) > 0:
    print("No root in this interval")
else:
    while abs(b-a) >= tol:
        xm = 0.5*(a+b)
        if f(xm) == 0:
            print("Root = ", xm)
            break
        elif f(a)*f(xm) < 0:
            b = xm
        else:
            a = xm

    print("Root = ", (a+b)*0.5)
```
