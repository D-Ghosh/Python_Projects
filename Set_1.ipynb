```python
# Q1(ii)
# Integration by composite Simpson's rule

import numpy as np
import math

a, b = eval(input("a, b\n"))
k = math.sqrt(1-(a**2/b**2))

def f(x): return 4*b*math.sqrt(1-(k**2)*math.sin(x)**2)

x1, x2, n = eval(input("Lower limit, Upper limit, n\n"))
x2 = x2/2
h = float(x2 - x1)/n
s = f(x1) + f(x2)
d = 4

for i in range(1, n):
    x = x1 + i*h
    s += d*f(x)
    d = 6-d

s = h/3*s
print(s)
```

```python
# Q2(ii)

def f(x): return (x**4)-1.99*(x**3)-1.76*(x**2)-5.22*x-2.33

a, b, tol = eval(input("Lower limit, Upper limit, Tolerance\n"))

if f(a)*f(b) > 0:
    print("No root in this interval")
else:
    while abs(b-a) >= tol:
        xm = 0.5*(a+b)
        if f(xm) == 0:
            print("Root = ", xm)
            break
        elif f(a)*f(xm) < 0:
            b = xm
        else:
            a = xm

    print("Root = ", (a+b)*0.5)
```

```python
# Q(ii)

import matplotlib.pyplot as plt
import math

def f(t, u): return (1-(4*t)/3)*u
def g(t): return math.exp(t-(2/3)*t**2)
u, t, h = 0, 1, 0.05
U, T = [], []

for i in range(1000):
    u += h*f(t, u)
    t += h
    T.append(t)
    U.append(u)

U2 = [g(i) for i in T]
plt.plot(T, U, T, U2)
plt.xlim(0, 5)
plt.ylim(0, 0.1)
plt.show()
```

```python
# Q7(ii)
# Lagrange Interpolation

x = np.arange(0, 2+0.2, 0.2)
y = np.sin(x**2)

m = len(x)
n = m-1 #Degree of polynomial
xp, yp = 0.9, 0

for i in range(m):
    p = 1
    for j in range(n+1):
        if j != i:
            p *= (xp-x[j])/(x[i]-x[j])
    yp += y[i]*p
print(yp)
```
